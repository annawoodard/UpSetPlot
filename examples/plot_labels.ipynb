```python
import matplotlib.pyplot as plt
from upsetplot import UpSet, from_contents
```

```python
contents = {
    'one': ['alpha', 'charlie', 'delta', 'echo', 'foxtrot', 'golf'],
    'two': ['delta', 'zulu'],
    'three': ['echo', 'golf', 'india', 'zulu'],
    'four': ['hotel', 'alpha', 'delta']
}

upset = UpSet(
    from_contents(contents),
    subset_size='count',
    intersection_plot_elements=3,
    sort_by='cardinality'
)

upset.plot()

plt.tight_layout()
plt.savefig('example.pdf')
```
