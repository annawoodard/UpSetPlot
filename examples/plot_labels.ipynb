```python
import matplotlib.pyplot as plt
from upsetplot import UpSet, from_contents
```

```python
contents = {
    'one': ['alpha', 'foxtrot', 'charlie', 'delta', 'echo', 'golf'],
    'two': ['delta', 'zulu'],
    'three': ['echo', 'golf', 'india', 'zulu'],
    'four': ['hotel', 'alpha', 'delta']
}

upset = UpSet(
    from_contents(contents),
    subset_size='count',
    intersection_plot_elements=3,
    sort_by='cardinality',
    plot_intersections=False
)

upset.plot()

plt.tight_layout()
plt.savefig('example_unsorted.pdf')
```

```python
upset = UpSet(
    from_contents(contents),
    subset_size='count',
    intersection_plot_elements=3,
    sort_by='cardinality',
    plot_intersections=False,
    element_order=[
        'golf', 'hotel', 'india', 'zulu', 'alpha', 'charlie', 'delta', 'echo', 'foxtrot'
    ]
)

upset.plot()

plt.tight_layout()
plt.savefig('example_sorted.pdf')
```
